```python
df_matchup = df[['eventID', 'gameID', 'visitor_or_home', 'inning', 'play_orig', 'batterID', 'pitcherID', 'pitch_count', 'pitch_sequence', 'theplay']]
df_matchup
```

```python
df_games = pd.read_csv('../datasets/raw/games.csv')
df_games.head()
```

```python
df_games.columns
```

```python
df_games = df_games[['gameID','date','dayofweek','visitor','home']]
df_games.head()
```

```python
df_teams = pd.read_csv('../datasets/raw/teams.csv')
df_teams.head()
```

```python
df_games
```

```python
df_games.columns
```

```python
df_games = pd.merge(df_games, df_teams, left_on='visitor', right_on='teamID', how='left')   
df_games['visitor_team_name'] = df_games[' city'] + ' ' + df_games[' nickname']
df_games = df_games.drop(columns=['teamID', ' league', ' city', ' nickname', ' yearStart', ' yearEnd'])
df_games.head()
```

```python
df_games = pd.merge(df_games, df_teams, left_on='home', right_on='teamID', how='left')   
df_games['home_team_name'] = df_games[' city'] + ' ' + df_games[' nickname']
df_games = df_games.drop(columns=['teamID', ' league', ' city', ' nickname', ' yearStart', ' yearEnd'])
df_games.head()
```

```python
df_merged = pd.merge(df_matchup, df_games, on='gameID', how='left')
df_merged.head()
```

```python
df_merged_after_2020 = df_merged[df_merged['date'] >= '2020-01-01']
df_merged_after_2020.head()
```
